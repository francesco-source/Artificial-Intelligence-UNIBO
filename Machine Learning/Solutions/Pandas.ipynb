```python
df["pussy_destroier"]=df["5.1"]*100
df.head()
```

```python
df["pussy_destroier"].mean()
df["pussy_destroier"].mode()
df["pussy_destroier"].std()
df["pussy_destroier"].mean()
df["pussy_destroier"].min()
df["pussy_destroier"].max()
df["pussy_destroier"].sum()
df["pussy_destroier"].quantile()
```

```python
def pt30(column):
    return column.quantile(0.3)
def pt40(column):
    return column.quantile(0.4)
df["pussy_destroier"].agg(pt30)
df[["5.1","pussy_destroier"]].agg(pt30)
df[["5.1","pussy_destroier"]].agg([pt30,pt40])
```

```python
df[["5.1","pussy_destroier"]].cumsum()#somma del primo elemento con quello 0, somma del secondo con quelli prima
```
