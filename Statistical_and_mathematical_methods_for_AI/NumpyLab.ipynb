### Linspace

Linspace(a,b,n) takes the first value of the vector (a), the last (b), and the number of elements in the vector (n). When he know this parameters it returns a vector filled with n elements equally separated one between each other from the first to the last.

```python
import matplotlib.pyplot as plt
a=0
b=1
n=50
x=np.linspace(a,b,n)
print(x)
x=np.linspace(0,2*np.pi,20)
y=np.sin(x)
print(y.shape)
plt.plot(x,y)
plt.title('A plot of f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('y = sin(x)')
plt.grid()
plt.show()
```

### Slicing operator and indicization

```python
x=np.array([1,2,3])
A=np.array([[1,2,3],[4,5,6]])
print(x)
print(x[1])
print(A)
print(A[1,2])
print(A[1][2])#this is c++, is correct but we don't do that here
print(x[1:])#this means print the elements from one to the end
print(x[1:3])#this means print the elements between one to 3-1
print(x[:2])#this means print all the elements  on the vector till the element 1
print(A[:,:2])# all the rows and the columns 0 and 1
```

### Matrix manipulations
1. A.T= transpose the matrix
1. np.linalg.matrix_rank(A)= gives the rank of the matrix
1. np.linalg.inv(A)= gives the inverse of A

```python
A=np.array([[1,2,3],[4,5,6],[7,8,8]])
print(A.T)
print (np.linalg.det(A))
print(np.linalg.matrix_rank(A))
print(np.linalg.inv(A))
print(np.linalg.inv(A)@A)# cause the computation is not exact the A^-1 A doesn't return the identity

```

### K_p(A)=||A||_p ||A^-1||_p conditonal number

```python
p=2#norm 2
print(np.linalg.cond(A,p))
p="fro"#frobeius norm
print(np.linalg.cond(A,p))
```

### Solving a linear system
Ax=y

```python
A=np.array([[1,2,3],[4,5,6],[7,8,8]])
y=np.array([0,1,0])
#Solve the system Ax=y
x_sol=np.linalg.solve(A,y)
print(f"the solution is{x_sol}")
```
